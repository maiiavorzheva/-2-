### ***Рассмотрим переменные***:

-**Клиент** - федеральная сеть, которая предоставила данные о третичных продажах

-**Сегмент** - название узкого сегмента внутри категории рынка Детской косметики (Семент < Категория)

-**Тип товара** - конкретный тип продукта/средства внутри сегмента (Тип продукта < Сегмент)

```python
# Для рассмотрения датасета в контексте центральных тенденций интересен показатель оффтейков
# Для этого определим среднее значение переменной ПРОДАЖИ В ШУТКАХ
ДетскаяКосметика['ПРОДАЖИ В ШТУКАХ'].mean()
```

```python
# Признак ряда, который лучше всего характеризует ценностную центральную тенденцию - это мода
# С помощью моды определим наиболее часто встречающееся значение варьирующего признака в вариационном ряду ПРОДАЖИ В ШТУКАХ.
ДетскаяКосметика['ПРОДАЖИ В ШТУКАХ'].mode()
```

Таким образом, большинство СКЮ на рынке Детской Косметики продавалось в кол-ве 1 шт. Это говорит о том, что оборот рынка генерится за счет наиболее продаваемых ТОП-СКЮ (см. пред. выводы).

## НЕПАРАМЕТРИЧЕСКИЕ КРИТЕРИИ

```python
import scipy.stats # импортируем библиотеку для расчета критерия Хи-квадрат
```

Сформулируем гипотезу: ТОП-вое СКЮ с большой долей вероятности сгенерило высокий оборот за счет продажи в Тандере.
Для этого перед тестированием создадим еще одну логическую переменную: Продажи в Тандере, где значение 0 будет говорит о том, что СКЮ не продавалось в Тандере, а 1 - что продавалось.

```python
ДетскаяКосметика['ПРОДАЖИ В ТАНДЕР'] = ДетскаяКосметика['КЛИЕНТ'].apply(lambda x: 1 if x == 'ТАНДЕР' else 0)
ДетскаяКосметика
```

Построим таблицу сопряженности по переменным ТОП-СКЮ и ПРОДАЖИ В ТАНДЕР

```python
ct_0 = ДетскаяКосметика.groupby('TOP-СКЮ')['ПРОДАЖИ В ТАНДЕР'].value_counts().unstack()
ct_0
```

По таблице сопряженности видим, что ТОПовые СКЮ продавались только в Тандере (в Ашане и Ленте - нет). Что касается СКЮ, не попавших в ТОП-рейтинг по обороту: 152 из них не продавались в Тандер, а 142 - продавались. 

```python
# посчитаем количество степеней свободы. Нас инетерсует (R-1)*(C-1)
dof = (2-1)*(2-1)
print(dof)
alpha = 0.01 #для нашей частотной таблицы статистический уровень значимости равен 0.01 
critical_value = 6.63 # а критическое значение для критерия Хи, при котором мы отвергнем нулевую гипотезу,  = 6.63
```

```python
n_total = ct_0.sum().sum()
print(n_total) # выведем все кол-во наблюдений (проверим, что оно соответствует кол-ву колонок в датасете)
```

```python
# просуммируем кол-во СКЮ, которые продавались в ТАНДЕРЕ и тех, которые не продавались там
n_tander_0 = ct_0[0].sum() # не продавались в Тандере
n_tander_1 = ct_0[1].sum() # продавались в Тандере
print(n_tander_0,n_tander_1)
```

```python
# рассчитаем вероятность того, что СКЮ продавалось в Тандере
p_tander_0 = n_tander_0 / n_total
p_tander_1 = n_tander_1 / n_total
print(p_tander_0,p_tander_1)
```

Таким образом, шанс того, что слаучайно выбранное СКЮ из датасета продавалось в Тандере - практически 50/50

```python
# теперь просчитаем кол-во СКЮ, которые попали в ТОП и кол-во СКЮ, которые не являются ТОПовыми
n_ordinary = ct_0.iloc[0].sum()
n_top = ct_0.iloc[1].sum()
print(n_ordinary,n_top) # обычные СКЮ vs. ТОПовые СКЮ
```

```python
# представляется возможным рассчитать теоретические частоты для обычных СКЮ, которые продавались в Тандере 
f_e_tander_0_ordinary = n_ordinary * p_tander_0
f_e_tander_1_ordinary = n_ordinary * p_tander_1
print(f_e_tander_0_ordinary,f_e_tander_1_ordinary)
```

Таким образом, ожидаем, что 147 обычных СКЮ не продавались в Тандере, и 146 - продавались

```python
# Продаелаем такое же упражнение для ТОПовых СКЮ, выясним, какое ожидаемое кол-во продается в Тандере, а какое- нет. 
f_e_tander_0_top = n_top * p_tander_0
f_e_tander_1_top = n_top * p_tander_1
print(f_e_tander_0_top,f_e_tander_1_top)
```

Таким образом, на основании выявленных ранее вероятностей, ожидаем, что 4 ТОПовых СКЮ не продавались в Тандере, и 3 - продавались

```python
# Построим матрицу теоретических (ожидаемых) частот 
ct_e = np.array([[f_e_tander_0_ordinary,f_e_tander_1_ordinary],
                [f_e_tander_0_top,f_e_tander_1_top]])
print(ct_e)
```

```python
# Теперь, когда имеем матрицу наблюдаемых частот и матрицу теоретических (ожидаемых) частот, можем вычислить критерий Пирсона 
chi_square = ((ct_0 - ct_e)**2 / ct_e).sum().sum()
print(chi_square)
```

```python
# Проверим, можем ли мы отвергнуть нулевую гипотезу
print(chi_square, critical_value)
print(chi_square > critical_value)
```

Таким образом, мы **не можем отвергнуть нулевую гипотезу** о том, что ТОПовые СКЮ сгенерили свой оборот за счет продаж в Тандере. Действительно, ТОПовые СКЮ в сентяьре 2022 года продавались только там.

# ЛИНЕЙНАЯ РЕГРЕССИЯ

```python
# импортируем все необходимые библиотеки
import seaborn as sns
import statsmodels.api as sm
```

```python
# Построим регрессионную модель для 2 зависимых переменных: ПРОДАЖИ В РУБЛЯХ и ПРОДАЖИ В ШТУКАХ
X = ДетскаяКосметика['ПРОДАЖИ В ШТУКАХ']
y = ДетскаяКосметика['ПРОДАЖИ В РУБЛЯХ']

res = scipy.stats.linregress(X, y)
print(res)
```

```python
# для полученной модели линейной регрессии рассчитаем коэффициент детерминации
r_sq = res.rvalue ** 2
print(r_sq)
```

Коэффициент детерминации имеет очень высокое значение, это говорит о том, что модель регрессии на 83,6% объясняет дисперсию (изменчивость) целевой переменной

```python
# Визуализируем результаты, отобразив линию регрессии на графике рассеяния для переменных ПРОДАЖИ В РУБЛЯХ И ШТУКАХ
plt.scatter(X, y, label = 'оригинальные данные')
plt.plot(X, res.intercept + res.slope * X, 'r', label = 'линия регрессии')
plt.legend()
plt.show()
```
